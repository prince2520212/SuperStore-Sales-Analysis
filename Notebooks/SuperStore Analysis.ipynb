```python
Q1 = eda_df['Sales'].quantile(0.25)
Q3 = eda_df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[
    (df['Sales'] < lower) |
    (df['Sales'] > upper)
]

outliers
```

```python
outlier_pct = (len(outliers) / len(eda_df)) * 100
print(f"Outliers: {outlier_pct:.2f}%")
```

# Finding Why Philadelphia is Generating Loss

```python
philly_loss = ( eda_df[eda_df['City'] == 'Philadelphia'] \
    .groupby('Sub-Category')['Profit'] \
    .sum() \
    .sort_values() \
    .head(10)
              )

plt.figure(figsize=(10,6))
philly_loss.plot(kind='barh')

plt.title("Top Loss-Making Sub-Categories in Philadelphia")
plt.xlabel("Profit")
plt.ylabel("Sub-Category")

plt.show()
```

```python
eda_df[eda_df['City'] == 'Philadelphia'] \
    .groupby('Sub-Category')['Discount'] \
    .mean() \
    .sort_values(ascending=False)
```

```python
philly = eda_df[eda_df['City'] == 'Philadelphia']

philly.groupby('Sub-Category')[['Sales','Profit','Discount']] \
      .mean() \
      .sort_values('Profit')
```

```python
print("Average Sales of Philadelphia : " ,philly.Sales.mean())
```

<h3 style="color:#d9534f;">Business Insight: Profitability Analysis for Philadelphia</h3>

<p>
Analysis of sales, profit, and discount data reveals that 
<b>Philadelphia's losses are primarily driven by aggressive discounting rather than weak demand.</b>
</p>

<ul>
    <li><b>Binders</b> and <b>Machines</b> receive average discounts of <b>70%</b>, resulting in substantial negative profitability.</li>
    <li><b>Phones</b> and <b>Tables</b> receive average discounts of <b>40%</b>, leading to losses despite generating strong sales.</li>
    <li><b>Bookcases</b> receive an average discount of <b>50%</b> and are among the largest contributors to overall losses.</li>
    <li>Several categories continue to generate revenue but fail to remain profitable due to excessive discounting.</li>
</ul>

<p>
<b>Suggestions:</b> Review discount policies for high-loss categories in Philadelphia and optimize pricing strategies to improve profitability while maintaining sales performance.
</p>

# MonthWise Sales Trend

```python
eda_df['Order Date'] = pd.to_datetime(eda_df['Order Date'])
eda_df['Month Group'] = eda_df['Order Date'].dt.month_name()

month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

eda_df['Month Group'] = pd.Categorical(
    eda_df['Month Group'],
    categories=month_order,
    ordered=True
)
```

```python
# Total Sales in months
monthly_sales = eda_df.groupby('Month Group' , observed=True)['Sales'].sum()

print(monthly_sales.sort_values())
plt.figure(figsize = (7,5))
sns.lineplot(monthly_sales)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

<h4 style="color:#28a745;">From the Graph November is the month with most sales follwed by September while February is the least</h4>

### MonthWise Profit 

```python
monthly_profit = eda_df.groupby('Month Group' , observed=True)['Profit'].sum()
print(monthly_profit.sort_values())
```

```python
plt.figure(figsize = (7,4))
plt.plot(monthly_profit.index , monthly_profit.values)
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.title("Monthly Profit")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

## Sales Vs Profit Trends

```python
sales_norm = monthly_sales / monthly_sales.max()
profit_norm = monthly_profit / monthly_profit.max()

plt.figure(figsize=(9,4))
plt.plot(monthly_sales.index , sales_norm, marker='o' , label='Sales')
plt.plot(monthly_profit.index , profit_norm , marker='s' ,linestyle='--', label='Profit')

plt.xlabel("Month")
plt.ylabel("Amount")
plt.title("Monthly Sales vs Profit")
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

<h4 style="color:#D12100;">From the Graph <b>November</b> is the month with most sales while <b>December</b> is the Most profitable</h4>
