```python
Q1 = eda_df['Sales'].quantile(0.25)
Q3 = eda_df['Sales'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers = df[
    (df['Sales'] < lower) |
    (df['Sales'] > upper)
]

outliers
```

```python
outlier_pct = (len(outliers) / len(eda_df)) * 100
print(f"Outliers: {outlier_pct:.2f}%")
```

 ### "Using the IQR method, 11.68% of records were identified as outliers in the Sales column. Upon inspection, these appear to represent legitimate high-value transactions rather than data quality issues, so they were retained for business analysis."

```python
philly_loss = ( eda_df[eda_df['City'] == 'Philadelphia'] \
    .groupby('Sub-Category')['Profit'] \
    .sum() \
    .sort_values() \
    .head(10)
              )

plt.figure(figsize=(10,6))
philly_loss.plot(kind='barh')

plt.title("Top Loss-Making Sub-Categories in Philadelphia")
plt.xlabel("Profit")
plt.ylabel("Sub-Category")

plt.show()
```

```python
eda_df[eda_df['City'] == 'Philadelphia'] \
    .groupby('Sub-Category')['Discount'] \
    .mean() \
    .sort_values(ascending=False)
```

```python
philly = eda_df[eda_df['City'] == 'Philadelphia']

philly.groupby('Sub-Category')[['Sales','Profit','Discount']] \
      .mean() \
      .sort_values('Profit')
```

```python
print("Average Sales of Philadelphia : " ,philly.Sales.mean())
```

<h3 style="color:#d9534f;">Business Insight: Profitability Analysis for Philadelphia</h3>

<p>
Analysis of sales, profit, and discount data reveals that 
<b>Philadelphia's losses are primarily driven by aggressive discounting rather than weak demand.</b>
</p>

<ul>
    <li><b>Binders</b> and <b>Machines</b> receive average discounts of <b>70%</b>, resulting in substantial negative profitability.</li>
    <li><b>Phones</b> and <b>Tables</b> receive average discounts of <b>40%</b>, leading to losses despite generating strong sales.</li>
    <li><b>Bookcases</b> receive an average discount of <b>50%</b> and are among the largest contributors to overall losses.</li>
    <li>Several categories continue to generate revenue but fail to remain profitable due to excessive discounting.</li>
</ul>

<p>
<b>Recommendation:</b> Review discount policies for high-loss categories in Philadelphia and optimize pricing strategies to improve profitability while maintaining sales performance.
</p>

```python
eda_df['Order Date'] = pd.to_datetime(eda_df['Order Date'])

eda_df['Month'] = eda_df['Order Date'].dt.month_name()
```

```python
month_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]

eda_df['Month'] = pd.Categorical(
    eda_df['Month'],
    categories=month_order,
    ordered=True
)
```

```python
monthly_sales = (
    eda_df.groupby('Month')['Sales']
    .sum()
    .sort_index()
)

monthly_sales
```

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(
    kind='line',
    marker='o'
)

plt.title('Month-wise Sales Trend')
plt.ylabel('Sales')
plt.grid(True)

plt.show()
```

<h3 style="color:#28a745;">ðŸ“ˆ Business Insight: Month-wise Sales Analysis</h3>

<p>
Sales exhibit a clear seasonal pattern throughout the year. Revenue remains relatively stable during most months but increases significantly during the fourth quarter.
</p>

<ul>
    <li><b>November</b> recorded the highest sales, generating over <b>350K</b> in revenue.</li>
    <li><b>February</b> was the weakest month, with sales dropping to approximately <b>60K</b>.</li>
    <li>Sales increased sharply in <b>March</b>, indicating a potential promotional event or increase in customer demand.</li>
    <li><b>Q4 (Octoberâ€“December)</b> consistently outperformed the rest of the year, suggesting strong seasonal demand during the holiday period.</li>
</ul>

<p>
<b>Recommendation:</b> Increase inventory availability and marketing efforts during Q4 to capitalize on seasonal demand. Additionally, investigate the factors driving March sales growth and replicate successful strategies in lower-performing months.
</p>

```python
monthly_profit = (
    eda_df.groupby('Month')['Profit']
    .sum()
)

monthly_profit
```

```python
plt.figure(figsize=(10,5))

monthly_profit.plot(
    kind='line',
    marker='*'
)

plt.title('Month-wise Profit Trend')
plt.ylabel('Profit')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker='o',
    label='Sales'
)

plt.plot(
    monthly_profit.index,
    monthly_profit.values,
    marker='s',
    label='Profit'
)

plt.title('Month-wise Sales and Profit Trend')
plt.xlabel('Month')
plt.ylabel('Amount')
plt.legend()
plt.grid(True)

plt.show()
```

```python
monthly = (
    eda_df.groupby('Month')[['Sales','Profit']]
    .sum()
    .reindex(month_order)
)

monthly['Profit Margin %'] = (
    monthly['Profit'] / monthly['Sales']
) * 100

monthly
```

```python
monthly['Profit Margin %'].plot(
    kind='bar',
    figsize=(12,5)
)

plt.title('Monthly Profit Margin (%)')
plt.ylabel('Profit Margin %')
plt.show()
```
